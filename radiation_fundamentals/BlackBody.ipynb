##Blackbody Radiation##
A perfect absorber ($a_\nu = 1$ at all
frequencies $\nu$) is called a black body. Kirchoff's law 
requires that every black body has emissivity $e_\nu = 1$ at all
frequencies as well, so the radiated spectrum $B_\nu(T)$ of any black
body at temperature $T$ is the same as the equilibrium spectrum of
radiation inside a cavity of temperature $T$, even if the walls of the
cavity are not black. Thus the intensity and spectrum of
blackbody radiation depends
only on the temperature of the black body or cavity. The same is
true for the electrical noise generated by a
warm resistor, a device that dissipates electrical energy, and which
plays an important role in radio
astronomy. You may already be familiar with the standard
derivations of the Rayleigh-Jeans and Planck radiation laws, but they
are
worth repeating because blackbody radiation is so fundamental and
because their one-dimensional analogs yield the spectrum
of electrical noise generated by a resistor.<br>

##The Rayleigh-Jeans Law##
Consider a large (meaning side length
$a \gg \lambda$, where $\lambda$ is the longest wavelength of interest)
cubical
** *cavity* **
filled with radiation in thermodynamic equilibrium. The purpose
of the cavity is to generate radiation and confine the radiation long
enough for it to
reach equilibrium. The radiation must be generated by thermal
accelerations of charged particles in the walls of any cavity with 
$T > 0$, but the equilibrium radiation is otherwise independent of the
wall material. We only require that the walls have nonzero conductivity
because walls of zero
conductivity would be transparent; they would generate no currents in
response to the electric
fields of incoming radiation and the radiation would pass through
unaffected. For walls having nonzero
conductivity, the transverse electric field strength at the walls is
$E = 0$ because modes with $E \ne 0$ at the walls are lossy. Only
those
standing waves with $E = 0$ at the walls will persist after some time
$t \gg
a/c$.

We can enumerate all possible
standing-wave modes in the cavity. For
example, consider all standing waves whose wave normals point in the
$x$
direction.<br>

<img src="http://www.cv.nrao.edu/course/astr534/images/swave.svg" height="243" width="481"><br>
*Standing waves*
corresponding to $n\lambda_{\rm x}/2 = a$ for $n = 1,~2,~3$.
Abscissa: x-axis of the cavity, bounded by walls at $x = 0$ and $x =
a$. Ordinate: Electric field strengths of the longest standing
waves satisfying the boundary condition $E = 0$ at the walls.<br>

---
The boundary conditions $E = 0$ at $x= 0$
and at $x = a$ mean that
only the waves having the discrete wavelengths<br>
$${\lambda_{\rm x} \over 2} = a, \qquad {2 \lambda_{\rm x} \over 2} =
a, \qquad {3 \lambda_{\rm x} \over 2} = a, ..., $$ <br>
can have nonzero amplitudes. Thus<br>
$${n \lambda_{\rm x} \over 2}= a$$ <br>
where $n = 1$, 2, 3, $...$. Similarly, <br>
$${n_{\rm x} \lambda_{\rm x} \over 2} = {n_{\rm y} \lambda_{\rm y} \over 2}= {n_{\rm z} \lambda_{\rm z} \over 2} = a$$ <br>
for wave normals in the $x,y,z$ directions. <br>

What about a wave whose normal is in
some arbitrary direction? Let
$\alpha,\,\beta,\,\gamma$ be the angles between the wave normal and the
$x,\,y,\,z$ axes, respectively. <br>

---
<img src="http://www.cv.nrao.edu/course/astr534/images/bbnodes.svg" height="383" width="449"><br>

This two-dimensional figure illustrates standing waves
propagating in a cavity with wave normals at angles $\alpha$ and $\beta$ from the $x$ and $y$ axes,
respectively. Examples of wave nodes, where $\vert E \vert = 0$, are indicated by dashed lines for the case $n_{\rm x} = 3$, $n_{\rm y}= 2$ .<br>

---
From the figure
it is clear that <br>
$$\lambda = \lambda_{\rm x} \cos\,
\alpha$$ <br>
where $\lambda$ is the wavelength measured in the direction of
the wave normal and $\lambda_{\rm x} \geq \lambda$ is the spacing
between the wave nodes
measured along the $x$-axis. Thus <br>
$$\lambda_{\rm x} = {\lambda \over \cos\,\alpha}$$ <br>
Similarly, <br>
$$\lambda_{\rm y} = {\lambda \over \cos\,\beta} <br>
{\rm ~~~and~~~} <br>
\lambda_{\rm z} = {\lambda \over \cos\,\gamma}$$ <br>
Our boundary conditions become <br>
$$n_{\rm x} = {2 a \over \lambda_{\rm x}},  n_{\rm y} = {2
a \over \lambda_{\rm
y}},  n_{\rm z} = {2 a \over \lambda_{\rm z}}$$ <br>
$$n_{\rm x} = {2 a \cos\,\alpha \over \lambda},  n_{\rm y}
= {2 a \cos\,\beta \over \lambda}, n_{\rm z} = {2 a \cos\,\gamma \over \lambda}$$ <br>
Squaring and summing the last three equations gives <br>
$$n_{\rm x}^2 + n_{\rm y}^2 + n_{\rm z}^2 = \biggl({2 a \over
\lambda}\biggr)^2 (\cos^2\,\alpha + \cos^2\,\beta + \cos^2\,\gamma)~.$$
<br>
The Pythagorean theorem implies $(\cos^2\,\alpha + \cos^2\,\beta +\cos^2\,\gamma) = 1$ so <br>
$$n_{\rm x}^2 + n_{\rm y}^2 + n_{\rm z}^2 = \biggl({2 a \over\lambda}\biggr)^2$$ <br>
The permitted frequencies $\nu = c / \lambda$ of these standing waves
are: <br>
$$\nu = {c \over 2 a}\sqrt{n_{\rm x}^2 + n_{\rm y}^2 + n_{\rm z}^2}$$ <br>
for all positive integers $n_{\rm x}$, $n_{\rm y}$, and $n_{\rm z}$.

The permitted standing-wave modes can
be represented as a lattice in the
positive octant of the space whose axes are $n_{\rm x}$, $n_{\rm y}$,
and $n_{\rm z}$. Each point of the lattice represents one possible mode
of equilibrium cavity radiation. The space density of points in this
lattice is unity, so the number of points in any volume is equal to
that volume. <br>

---
<img src="http://www.cv.nrao.edu/course/astr534/images/bboct.svg" height="373" width="382"><br>
*The (x,y) plane in
the imaginary
space whose axes are $(n_{\rm x}, \, n_{\rm y}, \, n_{\rm z})$.
Permitted standing waves in the $(n_{\rm x}, \, n_{\rm y})$ plane are
indicated by dots at positive integer values of these axes. *<br>

---
<br>
Let $\rho$ be the radial coordinate in $(n_{\rm x}, \, n_{\rm y}, \,
n_{\rm z})$-space. Then <br>
$$\rho^2 = n_{\rm x}^2 + n_{\rm y}^2 + n_{\rm z}^2$$ <br>
and <br>
$$\nu = {c \over \lambda} = {c \rho \over 2 a}$$ <br>
The number $N(\nu) d\nu$ of independent modes having frequencies in the
range $\nu$
to $\nu + d\nu$ equals the volume of the spherical octant shell between
$\rho$ and $\rho + d \rho$, multiplied by 2 to account for the two
independent polarizations of electromagnetic radiation: <br>
$$N(\nu) d\nu = {4 \pi \rho^2 d \rho \over 8} \times 2$$ <br>
$$N(\nu) d\nu = \pi \biggl({2 a \nu \over c}\biggr)^2 {2a \over c}
d\nu$$ <br>
In thermodynamic equilibrium at temperature $T$, equipartition of
energy implies that each mode has average
energy
$\langle E \rangle = kT$ according to the classical Boltzmann law (but
not
according to quantum mechanics). If we proceed using the
classical $kT$ for the
average energy per mode, the spectral energy density $u_\nu(T)$ of
cavity radiation in the frequency range $\nu$ to $\nu + d\nu$ is the
total
energy of all modes
in that frequency range divided by the volume $a^3$ of the cavity:<br>
$$
u_\nu(T) d\nu = {N(\nu) d\nu \over a^3} kT$$ <br>
$$u_\nu(T) = {8 \pi a^3 \over a^3} {\nu^2 \over c^3} kT = 8 \pi k T
{\nu^2 \over c^3}$$ <br>
The ** * spectral energy density * ** of radiation (blackbody or not) is spectral energy per
unit volume. It equals the total flow of spectral power per unit area
divided by the flow speed $c$: <br>
$$\bbox[border:3px blue solid,7pt]{u_\nu = {1 \over c} \int I_\nu d
\Omega}
\rlap{\quad \rm {(2C1)}}$$ <br>

Calling the specific intensity of blackbody radiation $B_\nu$ and
making use of the fact the blackbody radiation is isotropic, we
get for blackbody radiation: <br>
$$u_\nu = {1 \over c}\int_{4\pi} B_\nu d \Omega = {4 \pi \over c}
B_\nu$$ <br>
Thus <br>
$${4 \pi \over c} B_\nu = {8 \pi k T \nu^2 \over c^3}$$<br>
and we obtain the ** * Rayleigh-Jeans law* **
for the spectral brightness of blackbody radiation<br>
$$\bbox[border:3px blue solid,7pt]{B_\nu = {2 k T \nu^2 \over
c^2}}\rlap{\quad \rm {(2C2)}}$$<br>
that is valid only in the low-frequency limit $h \nu \ll k T$.<br>
 
Notice that:<br>
* Brightness is proportional to frequency squared because the volume of a spherical shell in three dimensions is proportional to frequency squared. <br>
* We assumed all modes have $\langle E \rangle = k T$, the classical assumption that breaks down at high frequencies.
* $B_\nu$ diverges at high $\nu$. This is called the "ultraviolet catastrophe".
* $B_\nu$ is independent of direction.</li>
* This result derived for a cubical cavity applies to equilibrium radiation in a cavity of any shape or (large) size, which you can demonstrate by a thought experiment in which a cubical cavity is connected through a small hole to the other cavity.

##The Planck Radiation Law##
The only flaw
in the derivation of
the Rayleigh-Jeans law is the
classical assumption that each radiation mode can have any energy $E$.
Then the continuous Boltzmann probability distribution <br>

$$P(E) \propto
\exp\biggl({-E \over k T}\biggr)$$ <br>
implies an average energy per mode<br>
$$\langle E \rangle ={\int_0^\infty E P(E) dE \over \int_0^\infty P(E) dE} = kT~.$$ <br>

Planck postulated that possible mode
energies are not
continuously distributed, but rather they are quantized and must
satisfy <br>
$$E = n h \nu, \qquad n = 1,~2,~3,~...$$ <br>
where $h \approx 6.63 \times
10^{-27}$ erg s is Planck's constant and $n$ is the number of ** *photons* **
(particles of light each having energy $E = h \nu$) in that
mode. Then <br>
$$P(E) = P(n h \nu) \propto
\exp\biggl({ - n h \nu \over k T}\biggr)$$ <br>
and the average energy per
mode is calculated by summing over only the discrete energies
permitted instead of integrating over all energies. <br>
$$\langle E \rangle =
{\sum_{n = 0}^\infty n h \nu P(n h \nu) \over \sum_{n = 0}^\infty P(n h
\nu)} = <br>
{\sum_{n = 0}^\infty n h \nu \exp\bigl({ - n h \nu \over k T}\bigr)
\over<br>
\sum_{n = 0}^\infty \exp\bigl({ - n h \nu \over k T}\bigr)}$$ <br>
Evaluating this sum yields <br>
$$ \langle E \rangle = {h \nu \over {\exp\bigl({h \nu \over kT}\bigr)
-1}} = k T \biggl[{h \nu / (k T) \over \exp [h \nu / (k T)] - 1}
\biggr]$$ <br>
Thus the
correct radiation law becomes <br>
$$B_\nu = {2 k T \nu^2 \over c^2}
\biggl[{h \nu / ( k T) \over \exp [h \nu / ( kT)] -1}
\biggr]~,$$ <br>
where the first factor is the Rayleigh-Jeans law and the quantity in
brackets is the quantum correction factor.<br>
** *Planck's law* ** 
 for blackbody
radiation is usually written in the form
z<br>
$$\bbox[border:3px blue<br>
solid,7pt]{B_\nu = {2 h \nu^3 \over c^2} <br>
{1 \over \exp\bigl({h \nu \over k T}\bigr)
-1}}<br>
\rlap{\quad \rm {(2C3)}}$$<br>
This eliminates the ultraviolet catastrophe because integrating over
all
frequencies gives a finite integrated brightness a blackbody radiator
at temperature $T$:
<br>
$$\bbox[border:3px blue<br>
solid,7pt]{B(T) \equiv \int_0^\infty B_\nu(T) d\nu = {\sigma
T^4 \over \pi}}<br>
\rlap{\quad \rm {(2C4)}}$$ <br>
where <br>
$$\sigma \equiv {2 \pi^5 k^4 \over 15 c^2 h^3} \approx 5.67 \times
10^{-5} {{\rm erg} \over {\rm cm}^2 {\rm ~s~K}^4 {\rm ~sr}}$$<br>
is called the ** *Stefan-Boltzmann
constant* **. Beware that the dimensionless sr in the
denominator is dropped by many authors.<br>

The spectral energy density per unit
solid angle of radiation is
just the specific intensity $I_\nu$ divided by the flow rate $c$.
For isotropic radiation, the spectral energy
density is<br>
$$u_\nu = 4 \pi
I_\nu / c$$ <br>
so the total ** *radiation
energy density* ** $u = \int_0^\infty u_\nu d \nu$ of
blackbody radiation is
<br>
$$\bbox[border:3px blue solid,7pt]{u = {4 \sigma T^4\over c}}<br>
\rlap{\quad \rm {(2C5)}}$$ The quantity $a \equiv 4 \sigma / c \approx
7.5657 \times 10^{-15} {\rm ~erg~cm}^{-3} {\rm ~K}^{-4}$ is called the **radiation
constant**.<br>

The frequency $\nu_{\rm max}$ at which
$B_\nu$, the brightness *per
unit frequency* of a
blackbody, is maximum is the solution of<br>
$$ {\partial B_\nu \over \partial \nu} = 0 ; $$<br>
it is<br>
$$ \biggl({\nu_{\rm max} \over {\rm GHz}}\biggr) \approx 59 \biggl( {T
\over {\rm K}}\biggr)~.$$<br>
Note that $\nu_{\rm max}$ is smaller than $c / \lambda_{\rm
max} \approx 103 {\rm ~GHz~} T {\rm (K)}$, where <br>
$$\biggl({\lambda_{\rm max} \over {\rm cm}}\biggr) \approx 0.29 \biggl(
{ {\rm K}\over
T} \biggr)$$ <br>
is the wavelength at which $B_\lambda$, the brightness *per unit wavelength*, is
highest. The latter equation is the familiar form of ** *Wien's Law* **
used by optical astronomers, whose spectrometers measure wavelengths instead of
frequencies.<br>

** The electrical power generated by
a warm resistor**

A resistor is any electrical device
that absorbs all of the
electrical power applied to it; it is the "black body" of electric
circuits. Motions of charged particles in a warm resistor at
temperature $T > 0$ K
generate electrical noise. The frequency spectrum of the
noise power depends only on the resistor temperature and is independent
of the resistor material. Also, the electrical noise generated is
indistinguishable from the noise coming from an antenna observing a blackbody radiator. Warm
resistors are useful in radio astronomy as standards
for calibrating receiver gains, and the power received by a radio telescope
is often described in terms of the "antenna temperature", the resistor
temperature required to generate the same power thermally. The gain of a radio receiver can be
calibrated by connecting it alternately to hot and cold resistors
(called "loads") having known temperatures, and the amount of noise
generated in a receiver can be described by the "receiver temperature",
the temperature of a resistor at the input of an imaginary noiseless
receiver having the same gain that would generate the same noise power
output. <br>

The derivation of the electrical power
per unit bandwidth $P_\nu$
generated by a resistor is a one-dimensional version of the
three-dimensional argument used for blackbody radiation [Nyquist, H.
1928, Phys. Rev. 32, 110; Barrett, A. H. 1970, Fundamentals of Radio
Astronomy, p. 22-23 (unpublished)]. At radio frequencies, it is often
true that $h\nu \ll
kT$ and the Rayleigh-Jeans limit applies. Recall that the
Rayleigh-Jeans
derivation of $B_\nu$ starts
with a large cube of side length $a$ containing standing waves of
thermal
radiation. The average energy in each standing-wave mode is $\langle E
\rangle = kT$, and
the number of modes with frequency $\nu$ to $\nu + d\nu$ is
proportional to $\nu^2$, so $B_\nu \propto \nu^2$. For the power
generated by a resistor,
consider two identical resistors at temperature $T$ connected by a
lossless transmission line (e.g., a pair of parallel wires) of length
$a$ much larger than the longest wavelength of interest. <br>

---
<img src="http://www.cv.nrao.edu/course/astr534/images/nyquist.svg" height="183" width="514"><br>
---
<br>
We assume that the transmission line
has a *characteristic impedance* equal to the resistance of the resistors so that
power
is
efficiently coupled between them. In equilibrium, the transmission line
can only support those standing waves having zero voltages at the ends;
other modes are suppressed by the lossy resistors. The surviving
standing waves satisfy<br>
$$ a = {n \lambda \over
2}, \qquad n = {\rm 1,~2,~3,}...,$$ <br>
where $\lambda$ is the wavelength.
Electrical signals do not travel at exactly the speed of light on a
transmission line, but at some slightly lower velocity $v < c$. Thus
<br>
$$n = {2 a \nu \over v}$$ <br>
and, for $a > \lambda$, the number of modes
per unit frequency is <br>
$${\Delta n \over \Delta \nu} = {2 a \over v}~.$$
<br>
In thermal equilibrium, the classical Boltzmann law says that each mode
has
average energy $\langle E \rangle = kT$, so the average energy per unit
frequency $E_\nu$
in the transmission line is <br>
$$E_\nu = {\Delta n \over \Delta \nu} kT =
{2akT \over v}$$ <br>
This energy takes a time $\Delta t = a / v$ to flow
from one end of the transmission line to the other, so the classical
power
(energy per unit time) per unit frequency flowing on the transmission
line is <br>
$$ P_\nu = {E_\nu \over \Delta t} = {2kT}$$ <br>

Note that the velocity $v$ has dropped
out. The total spectral
power $P_\nu$ generated by the two identical resistors must be $2kT$
and, by
symmetry, the spectral power generated by each resistor is
<br>
$$\bbox[border:3px blue solid,7pt]{P_\nu = kT}<br>
\rlap{\quad \rm {(2C6)}}$$<br>
in the limit $h \nu \ll k T$. This equation is called the **Nyquist
formula** and is the electrical
equivalent of the Rayleigh-Jeans law for radiation. Since the
"space" of the transmission line has only one dimension instead of
three, the frequency dependence is $\nu^0$ instead of $\nu^2$. <br>

Still, the Nyquist formula suffers
from an
"ultraviolet catastrophe" of its
own: the total power integrated over all frequencies diverges. Planck's
quantization rule is the cure for this problem also: the electrical
energy in
each mode must be an integer multiple of $h \nu$, and the corrected
version of the Nyquist formula is <br>
$$P_\nu= kT \biggl[{h \nu /( k T) \over \exp [h \nu / (kT)]-1}\biggr]~,$$<br>
where the quantity in large brackets is the same quantization
correction previously found for
blackbody radiation. This equation is usually written in the form
<br>
$$\bbox[border:3px blue solid,7pt]{P_\nu
= {h \nu \over {\exp\bigl({h \nu \over kT}\bigr)
-1}}}<br>
\rlap{\quad \rm {(2C7)}}$$ <br>

In many radio-astronomical applications, ${h \nu \ll kT}$ and the
simpler $P_\nu = k T$ (Eq. 2C6) can be used.<br>

---
<img src="http://www.cv.nrao.edu/course/astr534/images/bbspectrum.svg" height="414" width="396"><br>
At low
frequencies $\nu \ll k T / h$, the specific intensity $B_\nu$ of
blackbody radiation in three dimensions (solid curve) is proportional
to $\nu^2$. Its one-dimensional analog, the spectral power density
$P_\nu$ of noise generated by a resistor (dashed curve), is
proportional to $\nu^0$. Quantization causes both curves to cut
off exponentially at high frequencies.

```python
"""
This is the frequency version of blackbody enery density
"""

h=6.6244e-34 # Planck constant in Js
c=2.998e8     # speed of light in m/s
k=1.38e-23    #Boltzmann constant in J/K%

def blackbody(nu,T): # nu is frequency (Hz) , T is temperature (K)
    u_c=(8*pi*nu**2*k*T)/c**3 #classical part of energy density
    x=(h*nu)/(k*T)
    eb=x/(exp(x)-1 ) #Einstein-Bose correction factor - no units
    u_q=u_c*eb  # energy density
    S=u_q*c/4.0 # radiation
    return S
"""
 NOTE: for radiation per unit are, multiply u by c/4
- factor of 1/2 from radiation going in both directions
- factor of 1/2 from a cos(theta) factor intgrated over angle in 2D
"""

```

```python
for T in (3,100,300,1000,3000,5000,10000): #temperatures
    nu=logspace(6,14,140)#140 equally spaced points
    lb=log10(blackbody(nu,T))
    pylab.semilogx(nu,lb,label="%i K"%(int(T)))
    pylab.xlabel('freq(Hz)')
    pylab.ylabel('log flux per unit frequency emitted')
    pylab.legend(loc='upper left')
    pylab.title("Blackbody radiation")
    pylab.ylim(-30,-5)
```
