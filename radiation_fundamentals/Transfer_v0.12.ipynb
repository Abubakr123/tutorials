The brightness and flux density of a star at a set distanceWe distinguish between the brightness of the Sun, which does not depend on distance, and the apparent flux, which does.The quantitative definition of specific intensity or spectral brightness is$ I_{\nu} \equiv \frac{dP}{\cos{\theta}d\sigma d\Omega}$$d\sigma$ = infinitesimal surface area (e.g., of a detector)$\theta$ = the angle between a "ray" of radiation and the normal to the surface $d\sigma$ $d\Omega$ = infinitesimal solid angle measured at the observer's location

```python
Image(url="http://www.cv.nrao.edu/course/astr534/images/brtdef.svg")
```

If energy $dE_{\nu}$ flows through $d\sigma$ in time $dt$ in the frequency range to $\nu$ to   $\nu +d\nu$ within the solid angle $d\Omega$ on a ray which points an angle $\theta$ away from the surface normal, then the flow of energy is $dE_{\nu}=I_{\nu}cos(\theta) d\sigma d\Omega dt d\nu $The power $dP$ received in the solid angle $d\Omega$ and in the frequency range to $\nu$ to   $\nu +d\nu$ is $dP_{\nu}=I_{\nu}cos(\theta) d\sigma d\Omega d\nu  $Note : $dP_{\nu} = \frac{dE_{\nu}}{dt}$$\therefore$   $I_{\nu} \equiv \frac{dP_{\nu}}{cos(\theta) d\sigma d\Omega d\nu }$

##Theorem: Specific intensity is conserved (is constant) along any ray in empty space.##This follows directly from geometry.  Consider two surfaces, d1 and d2 along a ray of length r.

```python
Image(url='http://www.cv.nrao.edu/course/astr534/images/brtcon.svg')
```

Let $d\Omega_1$ be the solid angle subtended by  $d\sigma_2$ as seen from the center of the surface $d\sigma_1$.<br>Let $d\Omega_1$  be the solid angle subtended by  $d\sigma_1$ as seen from the center of the surface $d\sigma_2$.<br>Then<br>$d\Omega_1= \frac{cos\theta_2 d\sigma_2 }{r^2}$$d\Omega_2= \frac{cos\theta_1 d\sigma_1 }{r^2}$The energy $dW_1$ flowing through the area $d\sigma_1$  in solid angle  $d\Omega_1$ is<br>$dW_1=(I_{\nu})_1 cos\theta_1 d\Omega_1 d\sigma_1 d\nu$$dW_1=(I_{\nu})_1 cos\theta_1  \frac{cos\theta_2 d\sigma_2 }{r^2} d\sigma_1 d\nu$Likewise<br>$dW_2=(I_{\nu})_2 cos\theta_2  \frac{cos\theta_1 d\sigma_1 }{r^2} d\sigma_2 d\nu$Since the radiation energy is conserved in free space (where there is no absorption or emission), $dW_1=dW_2$, so<br>$(I)_1=(I)_2$This conservation theorem has two important consequences:(1) Brightness is independent of distance.  Thus the camera setting for a good exposure of the Sun would be the same, regardless of whether the photograph was taken close to the Sun (from near Venus, for example) or far away from the Sun (from near Mars, for example), so long as the Sun is resolved in the photograph. (2) Brightness is the same at the source and at the detector. Thus you can think of brightness in terms of energy flowing out of the source or as energy flowing into the detector.

The Total intensity is : $I \equiv \int_{0}^{\infty} I_{\nu}d\nu$  is also conserved.For sources that have a well-defined solid angle (not unresolved) $\frac {dP}{d\sigma d\nu} = I_{\nu} cos \theta d\Omega$Integrating of the sources extent gives us the flux density $S_{\nu}$ :$S_{\nu} \equiv \int_{source} I_{\nu}(\theta,\phi) cos \theta d\Omega \approx \int_{source} I_{\nu}(\theta,\phi) d\Omega  $for small angular sizes.
